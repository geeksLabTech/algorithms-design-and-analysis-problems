```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def compare_aproximation(data):
    
    sns.set_theme(style="whitegrid")

    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(6, 15))

    sns.set_color_codes("pastel")
    sns.barplot(x="BruteForce", y="Tests", data=data,
                label="Total", color="b", orient='h')

    sns.set_color_codes("muted")
    sns.barplot(x="Aproximation", y="Tests", data=data,
                label="Aproximation", color="b", orient='h')

    ax.legend(ncol=2, loc="lower right", frameon=True)
    ax.set(xlim=(0, 24), ylabel=""
    )
    sns.despine(left=True, bottom=True)
```

```python
data = pd.read_csv('dataframe.csv')
print(data)
compare_aproximation(data)
```
